### Selected Lane
Lane 2 – Refresh / Content Opportunity Scoring

I selected the Refresh / Content Opportunity Scoring lane because it focuses on identifying webpages that should be reviewed first for refresh, expansion, protection, pruning, or monitoring.

Large websites contain thousands of webpages, making manual review difficult and time-consuming. A machine learning model can analyze multiple page-level signals and prioritize the pages that are most likely to require attention.

The starter dataset contains useful metrics such as impressions, click-through rate (CTR), average search position, content age, and update history, making it suitable for this task.

### Research Question

Which webpages should be reviewed first for refresh, expansion, protection, pruning, or monitoring?

**Decision**
Determine which webpages should be prioritized for review.

**Unit of Analysis**
An individual webpage.

**Output**
A ranked review queue with a prediction score indicating priority.

**Action**
SEO teams review the highest-ranked webpages and decide whether to refresh, expand, protect, prune, or monitor them.

**Cost of Wrong Recommendation**

False Positive:
A healthy webpage is reviewed unnecessarily, wasting time and resources.

False Negative:
A declining webpage is overlooked, resulting in continued traffic loss and missed opportunities.

**Why Machine Learning?**

Machine learning can analyze multiple webpage metrics simultaneously and prioritize pages more effectively than simple manual rules.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nAverage CTR:")
print(round(df["ctr"].mean(), 4))

print("\nAverage Impressions:")
print(round(df["impressions_90d"].mean(), 2))

print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())
```

**Section 4: Careful words: what I can and can't claim**

The observations in this notebook are based only on the provided anonymized dataset.

The analysis identifies patterns within the available data but does not prove why webpages gain or lose traffic.

The machine learning model should be treated as a decision-support tool rather than a replacement for human decision making.

Any recommendations generated by the model should be reviewed before implementation.
